```python
import numpy as np
from tensorflow import keras
import tensorflow as tf
import tensorflow.keras
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import ModelCheckpoint, EarlyStopping, ReduceLROnPlateau
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, BatchNormalization, MaxPool2D, Flatten, Dense, Dropout
import cv2 as cv
import os
import pickle
import time
import random
import matplotlib.pyplot as plt

```

```python
scans = np.load("scans.npy",allow_pickle=True)
labels = np.load("labels.npy")
print(scans)
```

```python
Model = Sequential([Conv2D(filters=32,  kernel_size=(3,3), activation="relu", input_shape=scans.shape[1:]),
                    BatchNormalization(),
                    MaxPool2D(2,2),
                    Dropout(0.2),
                 
                    Conv2D(filters=16,  kernel_size=(3,3), activation="relu"),
                    BatchNormalization(),
                    MaxPool2D(3,3),
                    Dropout(0.2),
    
                    Flatten(),
    
                    Dense(units=8, activation="relu"),
                    BatchNormalization(),
                    Dropout(0.3),
                    
                                        
                    Dense(units=4, activation="relu"),
                    BatchNormalization(),
                    Dropout(0.5),
                    
                    Dense(2, activation='softmax'),
])

Model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

Model.summary()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(scans, labels, test_size=0.33)
Model.fit(X_train, y_train, epochs = 200)
```

```python
Model.save("final_79.h5")
model = keras.models.load_model("final_79.h5")
```

```python
def prepare(pic):
    pic = pic.reshape(-1, 200, 200,3)
    return pic/255


```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator,load_img
categories= ["N","S"]
def classify(image):
    try:
        frame = cv.imread(image)
        frame = prepare(frame)
    except:
        frame = image.reshape(-1,200,200,3)
    predictions = model.predict([frame])
    return predictions[0]
    
```

```python
image = X_test[9]
print(image.shape, y_test[9])
 
def predict(image):
    c = classify(image)
    p=0
    for a in c:
        if a > p: p = a
    return categories[(list(c).index(p))]
    
```
